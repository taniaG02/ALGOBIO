## Question 1: Finding the Longest Common Subsequence

We have not paid too much attention to it, but we have in the slides the basis of a dynamic programming algorithm to find the length of **longest common subsequence** between two strings, i.e. the longest, perhaps non consecutive, string that appears in both.  

Notice that **we do not impose** the constraint of the **characters in the sequence be consecutive**; with that requirement the problem is called the **common substring problem**.

Write first a Python function `lcs_matrix(s, t)` that receives two strings `s, t` and returns the dynamic programming matrix used to compute the length of their longest common subsequence.

Then write a Python function `lcs(s, t)` that returns a string with a longest common subsequence, using the LCS matrix returned by the previous function.

````python
import numpy as np

def lcs_matrix(s, t):
    """
    Creamos una matriz que tenga de dimensiones la longitud de cada palabra
    Input:
     - s: primera palabra
     - t: segunda palabra
    Output:
     - matrix: matriz
    """
    matrix = np.zeros(((len(t)+1), (len(s)+1)), int)
    for pos_letra_t in range (len(t)+1):
        for pos_letra_s in range (len(s)+1):
            if pos_letra_t != 0 and pos_letra_s!=0:
                if t[pos_letra_t-1] == s[pos_letra_s-1]:
                    matrix[pos_letra_t, pos_letra_s] = matrix[pos_letra_t-1, pos_letra_s-1] + 1
                else:
                    matrix[pos_letra_t, pos_letra_s] = max(matrix[pos_letra_t, pos_letra_s-1], matrix[pos_letra_t-1, pos_letra_s])

    return matrix


def lcs(s, t):
    """
    Utilizamos la matriz anterior y mmmm para encontrar la subsecuencia común utilizando la matriz que genera la función lcs_matrix.
    Input:
     - s: primera palabra
     - t: segunda palabra
    Output:
     - subseq: subsecuencia común.
    """
    subseq = ""
    matrix = lcs_matrix(s, t)
    pos_en_fila = len(s)  #Se refiere a la posición dentro de la fila
    for pos_letra_t in range (len(t), -1, -1):
        #print(pos_letra_t)
        if matrix[pos_letra_t, pos_en_fila] == 0: ## Terminará la función cuando sólo queden 0 en la matriz.
            break
        for pos_letra_s in range (pos_en_fila, -1, -1):
            #print(pos_letra_s)
            if matrix[pos_letra_t, pos_letra_s] == matrix[pos_letra_t, pos_letra_s-1]:
                continue
            elif matrix[pos_letra_t, pos_letra_s] == matrix[pos_letra_t-1, pos_letra_s]:
                break
            else:
                subseq += t[pos_letra_t-1]  # Vamos creando el string de la subsecuencia en orden inverso
                pos_en_fila=pos_letra_s ### Mejoramos haciendo que se detenga si s=0¿?
                break
    return subseq[::-1]  # Retornamos el string de la subsecuencia en el orden correcto
    

for s, t in zip(['bananas', 'biscuit', 'confidential'], ['bahamas', 'suitcase', 'trascendental']):
    lc_str = lcs(s, t)
    print(s, t, lc_str)
````

## Question 3: The 0-1 Knapsack Problem

As a sideline to our studies we are considering entering the bank robbing business, for which we must be able to solve the following problem: if we have a knapsack which stands a total integer weight of $W$ and there are $n$ items in the bank's vault with integer values $v_i$ and weights $w_i$, how do we choose those that maximize our loot without breaking the knapsack? Recall that we are in a 0-1 situation: you either take the entire item or none at all.

A greedy strategy (natural under the circumstances!) is to order the items by descending relative values $\frac{v_i}{w_i}$ and add them to the knapsack until the allowed total weight $W$ is surpassed. Remember that you cannot take cannot take a fraction of an item.

Write a function `greedy_value(l_weights, l_values, max_weight)` that returns the value of the maximal loot made up of elements with weights in `l_weights` and values in `l_values` that can be carried away in a knapsack which can hold at most a weight `max_weight`. 

````python
def greedy_value(l_weights, l_values, max_weight):
    """
    Esta función sigue una estrategia codiciosa para encontrar la combinación de l_weights que alcance el max_weight.
    Input:
     - l_weights: lista de pesos de los objetos.
     - l_values: lista de los valores de los objetos.
     - max_weight: Peso máximo del saco.
    Output:
     - ganancia: ganancia máxima que se obtendría
    """
    assert max_weight >= 0, "max_weight debe ser positivo." # Nos aseguramos de que el dato max_weight sea correcto (un valor positivo)
    val_rel = []                                            # Creamos una lista con los valores relativos para utilizar como índice próximamente.
    for order in range (len(l_weights)):
        val_rel.append(l_values[order]/l_weights[order])
    lista_orden_valores = list(zip(val_rel, l_weights, l_values))
    lista_orden_valores.sort(reverse = True)
    
    ganancia = 0
    for order in range (len(l_weights)):  
        objeto = lista_orden_valores[order]     # Creamos los valores relativos y el peso que necesitaremos en cada ciclo
        peso = objeto[1]
        ganancia += (max_weight // peso)*objeto[2]
        max_weight = max_weight % peso
        if max_weight == 0:         # Añadimos condición para que finalice el bucle si ya no queda más espacio en el saco 
                                    # (aunque para este caso, no es necesario).
            break
    return ganancia

l_weights = [4, 4, 5]
l_values  = [10, 11, 15]
max_weight = 8

print(greedy_value(l_weights, l_values, max_weight))
````

## Question 4: Dynamic Programming to Solve the 0-1 Knapsack Problem

Looking at the output of your greedy function on the previous example, you should be aware that your greedy strategy won't always give an optimal loot.
But don't worry, Dynamic Programming comes to the rescue!

Devise a dynamic programming strategy to maximize the loot and explain it in this markdown cell.

Then, write in the next cell a Python function `optimal_value(l_weights, l_values, max_weight)` that returns the value of the maximal loot made up of elements with weights in `l_weights` and values in `l_values` that can be carried away in a knapsack which can hold at most a weight `max_weight`. 

### Explicación de la estrategia de programación dinámica:

En esta función propongo trabajar con una lista de listas, formada cada una por el valor relativo, el peso y el precio (como en el ejercicio anterior). Con ello generaremos una matriz en cuyo eje "x" estarán los valores (números enteros) desde 0 hata el máximo de peso (llamado como capacidad) y en el eje "y" los valores del peso ordenados en función del valor relativo (de menor a mayor).

Por ello hacemos una matriz en la que,si el peso es mayor que la capacidad de esa columna, se pone el valor de la fila superior; si la canditad de peso indicada contiene el peso del objeto (una división de resto 0), el valor reflejado sea el de esa división y, si no es así, se ha de escoger la cantidad que otorgue más ganancias entre, en la misma columna, una fila superior; o el valor más una unidad que se encuentra tantas columnas atrás (como el peso) en la misma fila. 

Finalmente, se escoge el valor de la esquina inferior derecha y se recorre la matriz.

````python
def optimal_value(l_weights, l_values, max_weight):
    """
    Esta función utiliza pogramación dinámica para: mmmm
    Input:
     - l_weights: lista de pesos de los objetos.
     - l_values: lista de los valores de los objetos.
     - max_weight: Peso máximo del saco.
    Output:
     - ganancia: ganancia máxima que se obtendría
    """
    assert max_weight >= 0, "max_weight debe ser positivo." # Nos aseguramos de que el dato max_weight sea correcto (un valor positivo)
    # Primero preparamos la lista de valores relativos.
    val_rel = []
    for order in range (len(l_weights)):
        val_rel.append(l_values[order]/l_weights[order])
    lista_orden_valores = list(zip(val_rel, l_weights, l_values))
    lista_orden_valores.sort() # Tras ordenar los pesos, procedemos a la matriz

    # Después preparamos la matriz.
    matrix = np.zeros((len(lista_orden_valores), max_weight+1), int)
    sube = 0          # Preparamos una variable para restar en caso de que haya que ir a una fila superior a por un valor.
    for fila in range (len(lista_orden_valores)):
        peso = lista_orden_valores[fila][1]
        capacidad = 0 # Preparamos una variable para la posición de las columnas que se va a corresponder con el valor de la capacidad.
        while capacidad <= max_weight:
            if capacidad % peso == 0:
                matrix[fila, capacidad] = (capacidad // peso)  # Ponemos el cociente resultante
                if capacidad == (max_weight // peso):
                    sube = 0
            elif capacidad < peso:
                matrix[fila, capacidad] = matrix[fila - 1, capacidad]  # No hay que preocuparse de que el peso de las primeras líneas sea grande.
                                                                       # En la matriz se pondrá un 0 porque es el valor de las líneas que se indexarían.
            else:
                if (fila != 0) and ((capacidad) % lista_orden_valores[fila-sube][1]) == 0:  # Comprobamos que la fila no sea la primera y 
                                                # si el valor de la fila superior es de una división sin resto, para saber si este daría 
                                                # más ganancia que la cantidad de la fila actual.
                    if ((matrix[fila-sube, capacidad])*lista_orden_valores[fila-sube][2]) > ((capacidad // peso) * lista_orden_valores[fila][2]):
                        matrix[fila, capacidad] = matrix[fila-1, capacidad]  # En el caso en el que haya más ganancias en la fila anterior
                                                # que lo que daría el cociente de la fila actual, se conserva el valor de dicha fila
                else:
                    matrix[fila, capacidad] = (1 + matrix[fila, capacidad - peso])  
            capacidad += 1
        sube += 1

    # Finalmenterecorremos la matriz y obtenemos la ganancias.
    columna = (max_weight)
    fila = (len(l_weights)-1)
    ganancia = 0
    while columna > 0 and fila > 0:     # Seguir nuestras notas para la lectura de la matriz
        objeto = lista_orden_valores[fila]
        peso = objeto[1]
        if (matrix[fila, columna] == matrix[fila-1, columna]) and (columna % peso) != 0:
            fila -=1
        else:    
            cantidad = columna // peso
            ganancia += objeto[2]*cantidad
            columna -= peso*cantidad    #Las multiplicaciones por la cantidad tocaría quitarlas
            fila -=1
    return ganancia

l_weights = [4, 4, 5]
l_values  = [10, 11, 15]
max_weight = 8

print(optimal_value(l_weights, l_values, max_weight))
````
