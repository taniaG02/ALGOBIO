```python
from Bio import AlignIO
import logomaker
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el alineamiento (mejor práctica: usar 'with' para manejo de archivos)
try:
    alignment = AlignIO.read("clustalo-I20250321-165928-0743-92365021-p1m.fa", "fasta")
except FileNotFoundError:
    raise SystemExit("Error: Archivo de alineamiento no encontrado")

# Verificar que el alineamiento no esté vacío
if not alignment:
    raise ValueError("El alineamiento está vacío o no contiene secuencias")

# Extraer secuencias (versión optimizada)
sequences = [str(record.seq) for record in alignment]

# Verificar que todas las secuencias tengan la misma longitud
seq_lengths = {len(seq) for seq in sequences}
if len(seq_lengths) > 1:
    raise ValueError("Las secuencias del alineamiento tienen longitudes diferentes")

# Crear matriz de frecuencias (versión más eficiente)
frequency_matrix = logomaker.alignment_to_matrix(sequences)

# Alternativa manual (si prefieres control total):
# frequency_matrix = pd.DataFrame(
#     [pd.Series(list(col)).value_counts(normalize=True) 
#      for col in zip(*sequences)]
# ).fillna(0).T

# Crear y personalizar el logo
logo = logomaker.Logo(
    frequency_matrix,
    color_scheme='chemistry',  # Para proteínas (usar 'base_color' para ADN)
    font_name='Arial',
    show_spines=True,
    stack_order='big_on_top'
)

# Personalizar el gráfico
logo.style_spines(spines=['left', 'bottom'], visible=True)
logo.ax.set_xlabel('Posición en la secuencia', labelpad=5)
logo.ax.set_ylabel('Bits de información', labelpad=5)
logo.ax.set_title('Logo de secuencia generado a partir de ClustalW', pad=15)

# Mostrar el logo
plt.tight_layout()
plt.show()
```

```python
from Bio import AlignIO
import logomaker
import matplotlib.pyplot as plt
import numpy as np

# Cargar y procesar el alineamiento
alignment = AlignIO.read("clustalo-I20250321-165928-0743-92365021-p1m.fa", "fasta")
sequences = [str(record.seq) for record in alignment]

# Crear matriz de conteos (sin gaps)
counts_matrix = logomaker.alignment_to_matrix(sequences, to_type='counts', characters_to_ignore='-')

# Convertir a frecuencias relativas (ignorando gaps)
frequency_matrix = counts_matrix.div(counts_matrix.sum(axis=1), axis=0)

# Calcular información (en bits) para cada posición
information_matrix = logomaker.transform_matrix(frequency_matrix, from_type='probability', to_type='information')

# Crear el logo
logo = logomaker.Logo(
    information_matrix,  # Usamos la matriz de información directamente
    color_scheme='chemistry',
    font_name='Arial',
    show_spines=True,
    stack_order='big_on_top'
)

# Limitar bits a 4.32 si es necesario
logo.ax.set_ylim(0, 4.32)

# Personalización adicional
logo.ax.set_title("Logo de alineamiento ClustalW (excluyendo gaps)")
logo.ax.set_xlabel("Posición")
logo.ax.set_ylabel("Bits")

plt.tight_layout()
plt.show()
```
