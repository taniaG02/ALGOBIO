## Question 1: Finding the Longest Common Subsequence

We have not paid too much attention to it, but we have in the slides the basis of a dynamic programming algorithm to find the length of **longest common subsequence** between two strings, i.e. the longest, perhaps non consecutive, string that appears in both.  

Notice that **we do not impose** the constraint of the **characters in the sequence be consecutive**; with that requirement the problem is called the **common substring problem**.

Write first a Python function `lcs_matrix(s, t)` that receives two strings `s, t` and returns the dynamic programming matrix used to compute the length of their longest common subsequence.

Then write a Python function `lcs(s, t)` that returns a string with a longest common subsequence, using the LCS matrix returned by the previous function.

````python
def lcs_matrix(s, t):
    """
    Computes the Longest Common Subsequence (LCS) matrix for two given strings.

    Args:
        s (str): The first input string.
        t (str): The second input string.

    Returns:
        list[list[int]]: A 2D list representing the LCS matrix, where LCS[i][j] contains the length of the LCS of the substrings s[:i] and t[:j].

    Raises:
        AssertionError: If either input is not a string.   
    """
    
    # Function parameter control:
    assert isinstance(s, str), "First input must be a string."
    assert isinstance(t, str), "Second input must be a string."
    
    # If the strings are empty, there is no common subsequence.
    if len(s) == 0 or len(t) == 0:
        return ""
    
    # Handle any input type: convert everything to lowercase.
    s = s.lower()
    t = t.lower()
    
    # Initialize the matrix with zeros
    m, n = len(s), len(t)
    LCS = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Fill the matrix using dynamic programming
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s[i - 1] == t[j - 1]:
                LCS[i][j] = LCS[i - 1][j - 1] + 1  # Common character, add 1
            else:
                LCS[i][j] = max(LCS[i - 1][j], LCS[i][j - 1])  # Take the maximum of the cases
    
    return LCS


def lcs(s, t):
    """
    Uses the LCS matrix to find the longest common subsequence of two strings.

    Args:
        s (str): The first input string.
        t (str): The second input string.

    Returns:
        str: The longest common subsequence obtained by reading the "subsequence" list in reverse order.
    """
    # Get the LCS matrix
    LCS = lcs_matrix(s, t)
    
    # Reconstruct the longest common subsequence
    i, j = len(s), len(t)
    subsequence = []
    
    while i > 0 and j > 0:
        if s[i - 1] == t[j - 1]:
            subsequence.append(s[i - 1])  # If characters match, add to result
            i -= 1
            j -= 1
        elif LCS[i - 1][j] >= LCS[i][j - 1]:
            i -= 1  # Move up in the matrix
        else:
            j -= 1  # Move left in the matrix
    
    # Reverse the subsequence since it's built backwards
    return ''.join(reversed(subsequence))

for s, t in zip(['bananas', 'biscuit', 'confidential'], ['bahamas', 'suitcase', 'trascendental']):
    lc_str = lcs(s, t)
    print(s, t, lc_str)
````

````python
import numpy as np

def lcs_matrix(s, t):
    """
    Computes the Longest Common Subsequence (LCS) matrix for two strings using dynamic programming.

    Args:
        s (str): The first input string.
        t (str): The second input string.

    Returns:
        np.ndarray: A 2D NumPy array representing the LCS matrix.

    Raises:
        AssertionError: If either input is not a string.
    """
    # Function parameter control:
    assert isinstance(s, str), "First input must be a string."
    assert isinstance(t, str), "Second input must be a string."
    
    # If the strings are empty, there is no common subsequence.
    if len(s) == 0 or len(t) == 0:
        return np.zeros((len(t) + 1, len(s) + 1), dtype=int)
    
    # Handle any input type: convert everything to lowercase.
    s = s.lower()
    t = t.lower()
    
    # Initialize the matrix with zeros
    matrix = np.zeros((len(t) + 1, len(s) + 1), dtype=int)
    
    # Fill the matrix using dynamic programming
    for i in range(1, len(t) + 1):
        for j in range(1, len(s) + 1):
            if t[i - 1] == s[j - 1]:
                matrix[i, j] = matrix[i - 1, j - 1] + 1  # Common character, add 1
            else:
                matrix[i, j] = max(matrix[i, j - 1], matrix[i - 1, j])  # Take the maximum of the cases
    
    return matrix

def lcs(s, t):
    """
    Uses the LCS matrix to find the longest common subsequence of two strings.

    Args:
        s (str): The first input string.
        t (str): The second input string.

    Returns:
        str: The longest common subsequence.
    """
    # Get the LCS matrix
    matrix = lcs_matrix(s, t)
    col_idx = len(s)  # Column position in the matrix
    subseq = ""

    for row_idx in range(len(t), 0, -1):
        if matrix[row_idx, col_idx] == 0:  # Stop when only zeros remain in the matrix
            break
        for col in range(col_idx, 0, -1):
            if matrix[row_idx, col] == matrix[row_idx, col - 1]:
                continue
            elif matrix[row_idx, col] == matrix[row_idx - 1, col]:
                break
            else:
                subseq += t[row_idx - 1]  # Build the subsequence in reverse order
                col_idx = col  # Update column index
                break

    return subseq[::-1]  # Reverse the subsequence to return it in the correct order

for s, t in zip(['bananas', 'biscuit', 'confidential'], ['bahamas', 'suitcase', 'trascendental']):
    lc_str = lcs(s, t)
    print(s, t, lc_str)
````

````python
def pascal_row(k):
    """
    Computes the k-th row of Pascal's Triangle using recursion.

    Args:
        k (int): The index of the row in Pascal's Triangle to be returned. 
                 The 0th row is [1], the 1st row is [1, 1], etc.

    Returns:
        list: A list representing the k-th row of Pascal's Triangle.

    Raises:
        AssertionError: If the input is not a non-negative integer.
        AssertionError: If the calculated row does not follow Pascal's Triangle rule.
    """

    # Assert that the input argument is a non-negative integer
    assert isinstance(k, int) and k >= 0, "The value of k must be a non-negative integer."
        
    # Base case: the 0th row is [1]
    if k == 0:
        return [1]
    
    # Recursive for getting the previous row
    previous_row = pascal_row(k - 1)
    
    # Build the k-th row using the previous row
    current_row = [1]  # The first position is always 1
    for i in range(1, len(previous_row)):
        current_row.append(previous_row[i - 1] + previous_row[i])  # Sum of the previous elements
    current_row.append(1)  # The last position is always 1

    # Assert that the first and last elements are 1 (Pascal's Triangle property)
    assert current_row[0] == 1 and current_row[-1] == 1, "The calculated row does not follow Pascal's Triangle rule."
    
    return current_row

for k in range(1, 11):
    tr = pascal_row(k)
    print(tr)
````

````python
import numpy as np

def pascal_row_recursive(k):
    """
    Computes the k-th row of Pascal's Triangle using recursion with NumPy.

    Args:
        k (int): The index of the row in Pascal's Triangle to be returned.

    Returns:
        numpy.ndarray: A NumPy array representing the k-th row of Pascal's Triangle.

    Raises:
        AssertionError: If the input is not a non-negative integer.
        AssertionError: If the calculated row does not follow Pascal's Triangle rule.
    """
    # Assert that the input argument is a non-negative integer
    assert isinstance(k, int) and k >= 0, "The value of k must be a non-negative integer."

    # Base case: the 0th row is [1]
    if k == 0:
        return np.array([1])
    
    # Recursive case: get the previous row
    previous_row = pascal_row_recursive(k - 1)
    
    # Build the k-th row using the previous row
    new_row = np.zeros(k + 1, dtype=int)
    new_row[0], new_row[-1] = 1, 1
    new_row[1:-1] = previous_row[:-1] + previous_row[1:]

    # Assert that the first and last elements are 1 (Pascal's Triangle property)
    assert new_row[0] == 1 and new_row[-1] == 1, "The calculated row does not follow Pascal's Triangle rule."
    
    return new_row

for k in range(1, 11):
    tr = pascal_row(k)
    print(tr)
````

## Question 3: The 0-1 Knapsack Problem

As a sideline to our studies we are considering entering the bank robbing business, for which we must be able to solve the following problem: if we have a knapsack which stands a total integer weight of $W$ and there are $n$ items in the bank's vault with integer values $v_i$ and weights $w_i$, how do we choose those that maximize our loot without breaking the knapsack? Recall that we are in a 0-1 situation: you either take the entire item or none at all.

A greedy strategy (natural under the circumstances!) is to order the items by descending relative values $\frac{v_i}{w_i}$ and add them to the knapsack until the allowed total weight $W$ is surpassed. Remember that you cannot take cannot take a fraction of an item.

Write a function `greedy_value(l_weights, l_values, max_weight)` that returns the value of the maximal loot made up of elements with weights in `l_weights` and values in `l_values` that can be carried away in a knapsack which can hold at most a weight `max_weight`. 

````python
def greedy_value(l_weights, l_values, max_weight):
    """
    Esta función sigue una estrategia codiciosa para encontrar la combinación de l_weights que alcance el max_weight.
    Args:
     - l_weights: lista de pesos de los objetos.
     - l_values: lista de los valores de los objetos.
     - max_weight: Peso máximo del saco.
    Returns:
     - total_value: ganancia máxima que se obtendría.
    Raises:
     - ValueError: Si las listas de pesos y valores tienen diferente longitud.
     - ValueError: Si los pesos y los valores son números negativos.
     - ValueError: Si el peso máximo de la mochila es negativo.
     - TypeError: Si `l_weights` y `l_values` no son listas.
     - TypeError: Si`max_weight` no es un número entero.
    """
    # Metemos controles adecuados de los parámetros de la función:
    if len(l_weights) != len(l_values):
        raise ValueError("Las listas de pesos y valores deben tener la misma longitud.")
    if any(w < 0 for w in l_weights) or any(v < 0 for v in l_values):
        raise ValueError("Los pesos y los valores deben ser números no negativos.")
    if max_weight < 0:
        raise ValueError("El peso máximo de la mochila no puede ser negativo.")
    # Si cadenas vacias, devolver 0
    if len(l_weights) == 0 or len(l_values) == 0 or max_weight == 0:
        return 0

    if not isinstance(l_weights, list) or not isinstance(l_values, list):
        raise TypeError("`l_weights` y `l_values` deben ser listas.")
    if not isinstance(max_weight, int):
        raise TypeError("`max_weight` debe ser un número entero.")
    
    items = list(zip(l_weights, l_values))
    items.sort(key=lambda x: x[1] / x[0], reverse=True)
    
    total_value = 0
    total_weight = 0
    
    for weight, value in items:
        if total_weight + weight <= max_weight:
            total_weight += weight
            total_value += value
    
    return total_value

l_weights = [4, 4, 5]
l_values  = [10, 11, 15]
max_weight = 8

print(greedy_value(l_weights, l_values, max_weight))
````

## Question 4: Dynamic Programming to Solve the 0-1 Knapsack Problem

Looking at the output of your greedy function on the previous example, you should be aware that your greedy strategy won't always give an optimal loot.
But don't worry, Dynamic Programming comes to the rescue!

Devise a dynamic programming strategy to maximize the loot and explain it in this markdown cell.

Then, write in the next cell a Python function `optimal_value(l_weights, l_values, max_weight)` that returns the value of the maximal loot made up of elements with weights in `l_weights` and values in `l_values` that can be carried away in a knapsack which can hold at most a weight `max_weight`. 

````python
def optimal_value(l_weights, l_values, max_weight):
    """
    Esta función utiliza pogramación dinámica para: mmmm
    Args:
     - l_weights: lista de pesos de los objetos.
     - l_values: lista de los valores de los objetos.
     - max_weight: Peso máximo del saco.
    Returns:
     - Devuelve ganancia máxima que se obtendría, es el valor de la esquina inferior derecha de la matriz.
    Raises:
     - ValueError: Si las listas de pesos y valores tienen diferente longitud.
     - ValueError: Si los pesos y los valores son números negativos.
     - ValueError: Si el peso máximo de la mochila es negativo.
     - TypeError: Si `l_weights` y `l_values` no son listas.
     - TypeError: Si`max_weight` no es un número entero.
    """
    # Metemos controles adecuados de los parámetros de la función:
    if len(l_weights) != len(l_values):
        raise ValueError("Las listas de pesos y valores deben tener la misma longitud.")
    if any(w < 0 for w in l_weights) or any(v < 0 for v in l_values):
        raise ValueError("Los pesos y los valores deben ser números no negativos.")
    if max_weight < 0:
        raise ValueError("El peso máximo de la mochila no puede ser negativo.")
    # Si cadenas vacias, devolver 0
    if len(l_weights) == 0 or len(l_values) == 0 or max_weight == 0:
        return 0

    if not isinstance(l_weights, list) or not isinstance(l_values, list):
        raise TypeError("`l_weights` y `l_values` deben ser listas.")
    if not isinstance(max_weight, int):
        raise TypeError("`max_weight` debe ser un número entero.")
    
    n = len(l_weights)
    
    dp = [[0] * (max_weight + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, max_weight + 1):
            if l_weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - l_weights[i - 1]] + l_values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]
    
    return dp[n][max_weight]

l_weights = [4, 4, 5]
l_values  = [10, 11, 15]
max_weight = 8

print(optimal_value(l_weights, l_values, max_weight))
````
